```python
import requests
```

```python
from bs4
```

```python
response = requests.get("http://www.motherhood.com.my/diapers/?p=10")
response.status_code
```

```python
contents = bs4.BeautifulSoup(response.content,"html.parser")
```

```python
#getting pagenation
page_nation = contents.select(".pagination a")
```

```python
products = contents.select(".product-container")
```

```python

def checking(kinds, page):
    filtered = {}
    for i in range(page):
        response = requests.get("http://www.motherhood.com.my/{kinds}/?p={num}".format(kinds=kinds,num=i))
        contents = bs4.BeautifulSoup(response.content,"html.parser")

        for key, selected_product in enumerate(products):

            image_url = selected_product.select_one(".product_img_link").attrs["href"]
            item_url = selected_product.select_one(".product-name").attrs["href"]
            item_name = selected_product.select_one(".product-name").text.replace("\n", "").replace("\t","")
            price_list_raw = selected_product.select_one(".content_price").text.replace("\n", "").replace("\t","").split("RM")
            price = ' '.join(price_list_raw).split()
            filtered["{kinds}-{i}-{key}".format(kinds=kinds, i=i, key=key)] = {
                "image_url":image_url,
             "item_url":item_url,
             "item_name":item_name,
             "price":price,
             }
    return filtered
```

```python
checking("Food", 4)
```
